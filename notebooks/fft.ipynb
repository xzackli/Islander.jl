```julia
using Revise
using Islander
using PyPlot
using Traceur
using PyCall
using WCS
```

```julia
using PyCall
enmap = pyimport("pixell.enmap")

using FFTW
FFTW.set_num_threads(40)
```

```julia
wcs2 = WCSTransform(2;
    cdelt = [5/60., 5/60.],
    ctype = ["RA---CAR", "DEC--CAR"],
    crpix = [128.0, 64.0],
    crval = [0., 0.])
```

```julia
l, cl = Islander.util_get_cltt()
shape,wcs = enmap.geometry(shape=(128, 256),
                           res=deg2rad(5/60.),pos=(0,0))
modl = enmap.modlmap(shape, wcs);
cmb = enmap.rand_map(shape, wcs, 
    reshape(Islander.ellpad(cl),(1,1,length(cl)+1)));
noise = enmap.rand_map(shape, wcs, 1e-2 .* ones(1,1,30000))

cmb = Enmap(cmb, wcs2)
noise = Enmap(noise, wcs2);
```

```julia
fig, axes = plt.subplots(1, 2, figsize=(8,4))
axes[1].imshow(cmb)
axes[2].imshow(noise)
```

```julia
kcmb = enfft(cmb)
l_est, cl_est = Islander.ell_binpow( real(kcmb .* conj(kcmb)), modl, 50, 3000)
```

```julia
plot(l_est, cl_est, "ko", label="data" )
plot(l[1:3000], cl[1:3000], "r-", alpha=0.9, label="theory")
yscale("log")
legend()
```
